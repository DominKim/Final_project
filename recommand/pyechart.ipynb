```python
from pyecharts import options as opts
from pyecharts.charts import Bar, Grid, Line, Liquid, Page, Pie
from pyecharts.commons.utils import JsCode
from pyecharts.components import Table
from pyecharts.faker import Faker


def pie_cluster5() -> Pie:
    c = (
        Pie()
        .add(
            "",
            [list(z) for z in zip(df5_idx,df5_value)],
            center=["40%", "50%"],
            radius="65%",
        )
        .set_global_opts(
            title_opts=opts.TitleOpts(title="Cluster 5"),
            legend_opts=opts.LegendOpts(type_="scroll", pos_top = "20%", pos_left="80%", orient="vertical"),
        )
        .set_series_opts(label_opts=opts.LabelOpts(formatter="{b}: {c}"))
    )
    return c

def pie_cluster1() -> Pie:
    c = (
        Pie()
        .add(
            "",
            [list(z) for z in zip(df1_idx,df1_value)],
            center=["40%", "50%"],
            radius="65%",
        )
        .set_global_opts(
            title_opts=opts.TitleOpts(title="Cluster 1"),
            legend_opts=opts.LegendOpts(type_="scroll", pos_top = "20%", pos_left="80%", orient="vertical"),
        )
        .set_series_opts(label_opts=opts.LabelOpts(formatter="{b}: {c}"))
    )
    return c

def pie_cluster4() -> Pie:
    c = (
        Pie()
        .add(
            "",
            [list(z) for z in zip(df4_idx,df4_value)],
            center=["40%", "50%"],
            radius="65%",
        )
        .set_global_opts(
            title_opts=opts.TitleOpts(title="Cluster 4"),
            legend_opts=opts.LegendOpts(type_="scroll", pos_top = "20%", pos_left="80%", orient="vertical"),
        )
        .set_series_opts(label_opts=opts.LabelOpts(formatter="{b}: {c}"))
    )
    return c
def pie_cluster0() -> Pie:
    c = (
        Pie()
        .add(
            "",
            [list(z) for z in zip(df0_idx,df0_value)],
            center=["40%", "50%"],
            radius="65%",
        )
        .set_global_opts(
            title_opts=opts.TitleOpts(title="Cluster 0"),
            legend_opts=opts.LegendOpts(type_="scroll", pos_top = "20%", pos_left="80%", orient="vertical"),
        )
        .set_series_opts(label_opts=opts.LabelOpts(formatter="{b}: {c}"))
    )
    return c
def pie_cluster2() -> Pie:
    c = (
        Pie()
        .add(
            "",
            [list(z) for z in zip(df2_idx,df2_value)],
            center=["40%", "50%"],
            radius="65%",
        )
        .set_global_opts(
            title_opts=opts.TitleOpts(title="Cluster2"),
            legend_opts=opts.LegendOpts(type_="scroll", pos_top = "20%", pos_left="80%", orient="vertical"),
        )
        .set_series_opts(label_opts=opts.LabelOpts(formatter="{b}: {c}"))
    )
    return c
def pie_cluster7() -> Pie:
    c = (
        Pie()
        .add(
            "",
            [list(z) for z in zip(df7_idx,df7_value)],
            center=["40%", "50%"],
            radius="65%",
        )
        .set_global_opts(
            title_opts=opts.TitleOpts(title="Cluster7"),
            legend_opts=opts.LegendOpts(type_="scroll", pos_top = "20%", pos_left="80%", orient="vertical"),
        )
        .set_series_opts(label_opts=opts.LabelOpts(formatter="{b}: {c}"))
    )
    return c

def bar_graph() -> Bar:

    c = (
        Bar(init_opts=opts.InitOpts(theme=ThemeType.ROMA))
        .add_xaxis(["Cluster 4", "Cluster 0", "Cluster 5", "Cluster 1", "Cluster 2", "Cluster 7"])
        .add_yaxis("부채총계per자산총계", list2, stack="stack1", category_gap="50%")
        .add_yaxis("자본총계per자산총계", list3, stack="stack1", category_gap="50%")
        .set_global_opts(title_opts=opts.TitleOpts(title ="Cluster 별 부채총계per자산총계, 자본총계per자산총계"),
                        legend_opts=opts.LegendOpts(type_="scroll",  pos_left="80%", orient="vertical"))
        .set_series_opts(
            label_opts=opts.LabelOpts(
                position="right",
                formatter=JsCode(
                    "function(x){return Number(x.data.percent * 100).toFixed() + '%';}"
                ),
            )
        )
    )
    return c


405127


def page_simple_layout():
    page = Page(layout=Page.SimplePageLayout)
    page.add(
        pie_cluster4(),
        pie_cluster0(),
        pie_cluster5(),
        pie_cluster1(),
        pie_cluster2(),
        pie_cluster7(),
        bar_graph()
    )
    page.render("page_simple_layout.html")


if __name__ == "__main__":
    page_simple_layout()
```
